BGR ‚Üí RGB d√∂n√º≈ü√ºm√º yap

üëâ √ñƒüreneceƒüin:

```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("test.png")
print(img.shape)

img_bgr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(img_bgr.shape)

img_rgb = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

print(img_rgb.shape)

plt.imshow(img)
plt.imshow(img_bgr)
plt.imshow(img_rgb)
plt.axis("off")
```

```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("test.png")
print(img.shape)

roi = img[150:480, 700:1100]

img = cv2.rectangle(img, (690, 120), (1130, 480), (0, 255, 0), 2)
print(roi.shape)
plt.imshow(img)

plt.axis("off")
```

```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("test_blur.png")  # dosya adƒ±nƒ± d√ºzenle
titles = ["Original", "Bilateral"]
images = [
    img,
    cv2.bilateralFilter(img, 9, 150, 150)
]

plt.figure(figsize=(50,25))
for i in range(2):
    plt.subplot(1,5,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
plt.show()
```
